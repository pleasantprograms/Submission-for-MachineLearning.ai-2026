```python
import re

# ==========================
# 1. Franchise
# ==========================

def get_franchise(title):
    title = str(title)

    # Common two-word franchises
    if title.startswith("God of"):
        return "God of"
    if title.startswith("Devil May"):
        return "Devil May"
    if title.startswith("Call of"):
        return "Call of"
    if title.startswith("Need for"):
        return "Need for"
    if title.startswith("The Legend"):
        return "The Legend"
    if title.startswith("FIFA"):
        return "FIFA"

    # Otherwise first word
    return title.split()[0]

train_merged["franchise"] = train_merged["title"].apply(get_franchise)
test["franchise"] = test["title"].apply(get_franchise)


# ==========================
# 2. Year in title
# ==========================

train_merged["title_year"] = (
    train_merged["title"]
    .str.extract(r"(19\d{2}|20\d{2})")[0]
    .fillna("None")
)

test["title_year"] = (
    test["title"]
    .str.extract(r"(19\d{2}|20\d{2})")[0]
    .fillna("None")
)


# ==========================
# 3. Sequel Number
# ==========================

train_merged["sequel_number"] = (
    train_merged["title"]
    .str.extract(r"(\d+)$")[0]
    .fillna(0)
    .astype(int)
)

test["sequel_number"] = (
    test["title"]
    .str.extract(r"(\d+)$")[0]
    .fillna(0)
    .astype(int)
)


# ==========================
# 4. Sports Game
# ==========================

sports_keywords = [
    "FIFA","NBA","NFL","NHL",
    "MLB","PGA","F1",
    "Madden","WWE","UFC"
]

train_merged["is_sports"] = train_merged["title"].apply(
    lambda x: int(any(k in str(x) for k in sports_keywords))
)

test["is_sports"] = test["title"].apply(
    lambda x: int(any(k in str(x) for k in sports_keywords))
)


# ==========================
# 5. Word Count
# ==========================

train_merged["title_word_count"] = train_merged["title"].str.split().str.len()

test["title_word_count"] = test["title"].str.split().str.len()
```

```python
cat_features += [
    "title",
    "franchise",
    "title_year"
]
```

```python
duplicate_cols = []

cols = train_merged.columns

for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        c1 = cols[i]
        c2 = cols[j]

        if train_merged[c1].equals(train_merged[c2]):
            duplicate_cols.append((c1, c2))

print("Duplicate columns:")
for pair in duplicate_cols:
    print(pair)
```

```python
train_merged = train_merged.drop(columns=[
    "platform_generation_y",
    "publisher_tier_y"
])
```

```python
X = train_merged.drop('estimated_revenue_million_usd',axis='columns')
X.shape
```

```python
X
```

```python
X.marketing_campaign_type = X.marketing_campaign_type.fillna('None')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
cat_features = X_train.select_dtypes(include=['object','category']).columns.tolist()
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    loss_function='RMSE',
    random_seed=42,
    verbose=100,
    cat_features = cat_features
)

model.fit(
    X_train,
    y_train_log,
    cat_features=cat_features
)
```

```python
catboost_pred_log = model.predict(X_test)
catboost_pred = np.expm1(catboost_pred_log)
print(y_test)
print(catboost_pred)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    root_mean_squared_error,
    r2_score,
    root_mean_squared_log_error
)
```

```python
print("CatBoost")
print("MAE :", mean_absolute_error(y_test, catboost_pred))
print("MSE :", mean_squared_error(y_test, catboost_pred))
print("RMSE:", root_mean_squared_error(y_test, catboost_pred))
print("R²  :", r2_score(y_test, catboost_pred))
print("RMSLE :", root_mean_squared_error(y_test_log,catboost_pred_log))
```
