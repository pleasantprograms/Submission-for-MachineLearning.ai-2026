```python
y = train_dummy.estimated_revenue_million_usd
y
```

```python
X = train_dummy.drop(['internal_build_id','title','game_id','estimated_revenue_million_usd'],axis='columns')
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    root_mean_squared_error,
    r2_score,
    root_mean_squared_log_error
)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import ExtraTreesRegressor
from xgboost import XGBRegressor
```

```python
xgb = XGBRegressor()
```

```python
xgb.fit(X_train,y_train_log)
```

```python
xgb_pred_log = xgb.predict(X_test)
```

```python
xgb_pred = np.expm1(xgb_pred_log)
```

```python
print("XGB")
print("MAE :", mean_absolute_error(y_test, xgb_pred))
print("MSE :", mean_squared_error(y_test, xgb_pred))
print("RMSE:", root_mean_squared_error(y_test, xgb_pred))
print("R²  :", r2_score(y_test, xgb_pred))
```

```python
print(y_test[:10])
print(xgb_pred[:10])
```

```python
y_test_log = np.log1p(y_test)
```

```python
from sklearn.metrics import root_mean_squared_error

rmsle = root_mean_squared_error(y_test_log, xgb_pred_log)
print("RMSLE:", rmsle)
```

```python
from sklearn.model_selection import RandomizedSearchCV

xgb = XGBRegressor(
    objective='reg:squarederror',
    random_state=42,
    tree_method='hist',      # Much faster than 'exact'
    n_jobs=-1
)

param_dist = {
    'n_estimators': [200, 300, 400],
    'max_depth': [4, 5, 6, 7],
    'learning_rate': [0.03, 0.05, 0.1],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.3],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [1, 2, 5]
}

random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_dist,
    n_iter=20,              # Only 20 combinations
    scoring='neg_root_mean_squared_error',
    cv=3,                   # 3-fold CV
    verbose=2,
    random_state=42,
    n_jobs=-1
)

X_tune, _, y_tune, _ = train_test_split(
    X_train,
    y_train,
    train_size=0.5,
    random_state=42
)
y_tune_log = np.log1p(y_tune)
random_search.fit(X_tune, y_tune_log)
```

```python
best_xgb = XGBRegressor(
    objective='reg:squarederror',
    random_state=42,
    tree_method='hist',
    n_jobs=-1,
    **random_search.best_params_
)

best_xgb.fit(X_train, y_train_log)
```

```python
xgb_pred_log = best_xgb.predict(X_test)
xgb_pred = np.expm1(xgb_pred_log)
print("XGB")
print("MAE :", mean_absolute_error(y_test, xgb_pred))
print("MSE :", mean_squared_error(y_test, xgb_pred))
print("RMSE:", root_mean_squared_error(y_test, xgb_pred))
print("R²  :", r2_score(y_test, xgb_pred))
```

```python
y_test_log = np.log1p(y_test)
rmsle = root_mean_squared_error(y_test_log, xgb_pred_log)
print("RMSLE:", rmsle)
```
