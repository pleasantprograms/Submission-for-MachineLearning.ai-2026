```python
train.marketing_campaign_type = train.marketing_campaign_type.fillna('None')
train.isnull().sum()
```

```python
X = train.drop('estimated_revenue_million_usd',axis='columns')
y = train.estimated_revenue_million_usd
```

```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
import re

KNOWN_FRANCHISES = [
    "Call of Duty",
    "The Witcher",
    "Dark Souls",
    "Final Fantasy",
    "God of War",
    "No Man's Sky",
    "Red Dead Redemption",
    "Grand Theft Auto",
    "MLB The Show",
    "Need for Speed",
    "Rainbow Six",
    "Counter-Strike",
    "Assassin's Creed",
    "Ghost Recon",
    "Metal Gear",
    "Devil May Cry",
    "Super Mario",
    "Batman",
    "Halo",
    "FIFA",
    "NBA",
    "NHL",
    "Madden",
    "Overwatch",
    "Battlefield",
    "Titanfall",
    "Minecraft",
    "Doom",
    "UFC",
    "F1",
    "WWE",
    "eFootball"
]
```

```python
def extract_franchise(title):
    title_lower = title.lower()

    # Match known multi-word franchises first
    for franchise in sorted(KNOWN_FRANCHISES, key=len, reverse=True):
        if title_lower.startswith(franchise.lower()):
            return franchise

    # Remove years
    title = re.sub(r"\b(19|20)\d{2}\b", "", title)

    # Remove standalone sequel numbers
    title = re.sub(r"\b\d+\b", "", title)

    # Remove edition words
    remove_words = {
        "Remastered",
        "Ultimate",
        "Definitive",
        "Edition",
        "Complete",
        "Collection",
        "Evolved",
        "Reloaded",
        "Enhanced"
    }

    words = [w for w in title.split() if w not in remove_words]

    # Fallback: first word
    return words[0].strip(":")
```

```python
sample = [
    "FIFA 2025",
    "Overwatch 6",
    "Batman: Arkham Valley",
    "The Witcher 2: Reckoning",
    "No Man's Sky Evolved",
    "Call of Duty Black Ops",
    "Dark Souls III",
    "MLB The Show 2026"
]

for s in sample:
    print(s, " --> ", extract_franchise(s))
```

```python
X_train["franchise"] = X_train["title"].apply(extract_franchise)
```

```python
print(X_train["franchise"].value_counts().head(30))
```

```python
X_test["franchise"] = X_test["title"].apply(extract_franchise)
```

```python
cat_features = X_train.select_dtypes(include=['object','category']).columns.tolist()
cat_features
```

```python
y_test_log = np.log1p(y_test)
y_train_log = np.log1p(y_train)
y_train_log.shape
```

```python
model1 = CatBoostRegressor(
    iterations=600,
    learning_rate=0.05,
    depth=6,
    loss_function='RMSE',
    random_seed=42,
    verbose=100,
    cat_features = cat_features
)

model1.fit(
    X_train,
    y_train,
    cat_features=cat_features
)
```

```python
model2 = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    loss_function='RMSE',
    random_seed=42,
    verbose=100,
    cat_features = cat_features
)

model2.fit(
    X_train,
    y_train_log,
    cat_features=cat_features
)
```

```python
model1_pred = model1.predict(X_test)
model1_pred = np.clip(model1_pred,0,None)
model1_pred_log = np.log1p(model1_pred)
print(y_test[:20])
print(model1_pred[:20])
print(y_test_log[:20])
print(model1_pred_log[:20])
```

```python
model2_pred_log = model2.predict(X_test)
model2_pred = np.expm1(model2_pred_log)
print(y_test[:20])
print(model2_pred[:20])
print(y_test_log[:20])
print(model2_pred_log[:20])
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    root_mean_squared_error,
    r2_score,
    root_mean_squared_log_error
)
```

```python
print("CatBoost 1")
print("MAE :", mean_absolute_error(y_test, model1_pred))
print("MSE :", mean_squared_error(y_test, model1_pred))
print("RMSE:", root_mean_squared_error(y_test, model1_pred))
print("R²  :", r2_score(y_test, model1_pred))
print("RMSLE :", root_mean_squared_error(y_test_log,model1_pred_log))
```

```python
print("CatBoost 2")
print("MAE :", mean_absolute_error(y_test, model2_pred))
print("MSE :", mean_squared_error(y_test, model2_pred))
print("RMSE:", root_mean_squared_error(y_test, model2_pred))
print("R²  :", r2_score(y_test, model2_pred))
print("RMSLE :", root_mean_squared_error(y_test_log,model2_pred_log))
```

```python
from sklearn.metrics import root_mean_squared_log_error

model2_pred = np.clip(model2_pred, 0, None)
score = root_mean_squared_log_error(y_test, model2_pred)
print(score)
```

```python
final_pred = 0.5*model1_pred + 0.5*model2_pred
final_pred_log = np.log1p(final_pred)
```

```python
print("Final")
print("MAE :", mean_absolute_error(y_test, final_pred))
print("MSE :", mean_squared_error(y_test, final_pred))
print("RMSE:", root_mean_squared_error(y_test, final_pred))
print("R²  :", r2_score(y_test, final_pred))
print("RMSLE :", root_mean_squared_error(y_test_log,final_pred_log))
```
